```python
df = pd.read_csv("data.csv")
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Suponiendo que 'SEMANA', 'VENTA', y 'STOCK' son columnas de tu DataFrame 'df'
fecha = df['SEMANA']
valores = df['VENTA']
valores_linea = df['STOCK']

# Convertir las fechas a nmeros de punto flotante para usar en el grfico
fecha_numerica = np.arange(len(fecha))

# Agregar un pequeo desplazamiento aleatorio a las posiciones de las fechas
desplazamiento = np.random.uniform(low=-0.1, high=0.1, size=len(fecha))
fecha_desplazada = fecha_numerica + desplazamiento

# Crear la figura y los ejes
fig, ax1 = plt.subplots(figsize=(20, 6))

# Crear las barras en el eje y1
barras = ax1.bar(fecha_desplazada, valores, color='blue')
ax1.set_ylabel('Barras', color='blue')

# Crear un segundo eje y2 para la lnea
ax2 = ax1.twinx()
linea = ax2.plot(fecha_desplazada, valores_linea, color='red', marker='o', label='Lnea')
ax2.set_ylabel('Lnea', color='red')

# Establecer etiquetas y nombres de semanas en el eje x
ax1.set_xticks(fecha_desplazada)
ax1.set_xticklabels(fecha, rotation=90)

# Agregar etiquetas en las barras
for barra, valor in zip(barras, valores):
    ax1.text(barra.get_x() + barra.get_width()/2, barra.get_height(), str(valor), ha='center', va='bottom')

# Agregar etiquetas en los puntos de la lnea
for x, y in zip(fecha_desplazada, valores_linea):
    ax2.text(x, y, str(y), ha='center', va='bottom')

# Configuraciones adicionales
ax1.set_xlabel('Semana')
ax1.set_title('Grfica de barras y lnea')
ax2.legend()

# Mostrar la grfica
plt.tight_layout()
plt.show()
```
